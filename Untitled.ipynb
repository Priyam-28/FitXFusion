```python
import matplotlib.pyplot as plt
import numpy as np

exercise_types = ['Squats', 'Lunges', 'Push-ups', 'Pull-ups', 'Bicep curls',
                  'Shoulder presses', 'Deadlifts', 'Planks', 'Jumping jacks', 'Burpees',
                  'Mountain climbers', 'Russian twists', 'Leg raises', 'Crunches', 'Dips']
accuracy = [0.95, 0.92, 0.94, 0.90, 0.93, 0.89, 0.91, 0.97, 0.96, 0.92,
            0.90, 0.88, 0.91, 0.93, 0.90]

plt.figure(figsize=(12, 6))
plt.bar(exercise_types, accuracy)
plt.title('Classification Accuracy by Exercise Type')
plt.xlabel('Exercise Type')
plt.ylabel('Accuracy')
plt.ylim(0.8, 1.0)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

exercise_types = ['Squats', 'Lunges', 'Push-ups', 'Pull-ups', 'Bicep curls']
correct_form = [75, 70, 65, 60, 72]
incorrect_form = [25, 30, 35, 40, 28]

x = np.arange(len(exercise_types))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))
rects1 = ax.bar(x - width/2, correct_form, width, label='Correct Form', color='g')
rects2 = ax.bar(x + width/2, incorrect_form, width, label='Incorrect Form', color='r')

ax.set_ylabel('Percentage')
ax.set_title('Form Analysis: Correct vs. Incorrect Form')
ax.set_xticks(x)
ax.set_xticklabels(exercise_types)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)

fig.tight_layout()
plt.show()
```
